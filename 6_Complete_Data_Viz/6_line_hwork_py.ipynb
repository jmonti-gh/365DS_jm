```python
# Import/read the data
df_returns_10_20 = pd.read_csv('6_returns1020.csv')
df_returns_10_20
```

```python
print(df_returns_10_20.info())
df_returns_10_20.Date.dtype
```

```python
df_returns_10_20['dt_date'] = pd.to_datetime(df_returns_10_20['Date'])
display(df_returns_10_20.iloc[[0, 4, 9, -9, -5, -1]])
df_returns_10_20.dtypes
```

```python
plt.figure(figsize=(20, 6))

plt.plot(df_returns_10_20['dt_date'], df_returns_10_20['GSPCRet'])
plt.plot(df_returns_10_20['dt_date'], df_returns_10_20['FTSERet'])

plt.title('S&P 500 vs. FTSE 100 stoks returns (2010 - 2020)')
plt.legend(['S&P 500', 'FTSE 100'])

plt.show()
```

### viewing a narrow period (H2 2012)

```python
df_returns_h2_2012 = df_returns_10_20[(df_returns_10_20['dt_date'] >= '2012-01-01') &
                                      (df_returns_10_20['dt_date'] <= '2012-12-31')]
df_returns_h2_2012
```

```python
plt.figure(figsize=(20, 6))

plt.plot(df_returns_h2_2012['dt_date'], df_returns_h2_2012['GSPCRet'], color='midnightblue')
plt.plot(df_returns_h2_2012['dt_date'], df_returns_h2_2012['FTSERet'], color='crimson')

plt.title('S&P 500 vs. FTSE 100 stoks returns (2010 - 2020)')
plt.legend(['S&P 500', 'FTSE 100'])

plt.show()
```

### Make a function which show the period passed as argument.

```python
def lc_period(start, end):
    df_returns_period = df_returns_10_20[(df_returns_10_20['dt_date'] >= start) &
                                         (df_returns_10_20['dt_date'] <= end)]
    plt.figure(figsize=(20, 6))
    plt.plot(df_returns_period['dt_date'], df_returns_period['GSPCRet'], color='midnightblue')
    plt.plot(df_returns_period['dt_date'], df_returns_period['FTSERet'], color='crimson')
    plt.title(f'S&P 500 vs. FTSE 100 stocks returns ({start} - {end})')
    plt.legend(['S&P 500', 'FTSE 100'])
    plt.show()
    
```

```python
lc_period('2015-07-01', '2015-12-31')
```
