```python
# Import/read the data
df_scatter = pd.read_csv('8_student_scores_data.csv')
df_scatter
```

```python
print(df_scatter.info())
```

```python
# Matplotlib Scatter Plot
plt.figure(figsize=(10, 6))

scatter = plt.scatter(df_scatter['SAT'],       # x-axis (var 1)
                      df_scatter['GPA'],            # y-axis (var 2)
                      alpha=0.6,                        # Transparency
                      cmap='viridis')                   # Palette

```

```python
# Seaborn sns.scatterplot()
plt.figure(figsize=(9, 6))
sns.scatterplot(x=df_scatter['SAT'],
                y=df_scatter['GPA'],
                s=50)
```
