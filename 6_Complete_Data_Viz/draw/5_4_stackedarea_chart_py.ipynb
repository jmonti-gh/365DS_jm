## 5.- Stacked Chart in python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(pd.__version__)
print(sns.__version__)

sns.set()               # this is enough to beauty the chart
```

```python
# Read / import the data
df = pd.read_csv('stacked_area_chart_data.csv')
df
```

```python
# Simple matplotlib stackedarea chart
xy_size = 13
colors_hex = ['#011638', '#7e2987', '#ef2026']
ordered_cols = ['Diesel', 'Petrol', 'Gas']
sns.set_style('white')

plt.figure(figsize=(10, 6))

plt.stackplot(df['Year'],
        df[ordered_cols[0]],
        df[ordered_cols[1]],
        df[ordered_cols[2]],
        colors=colors_hex,
        edgecolor='none')

plt.title('Cars Sold by Fuel Engine Type', fontsize=16, fontweight='bold')
plt.xticks(df['Year'], rotation=45)
plt.ylabel('Number of cars', fontsize=xy_size)
plt.legend(ordered_cols)
sns.despine()
plt.show()

# Note: general palette -> sns.set_palette('colorblind')
```
