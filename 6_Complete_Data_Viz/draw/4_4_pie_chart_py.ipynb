## 4.- Pie Chart in python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(pd.__version__)
print(sns.__version__)

sns.set()               # this is enough to beauty the chart
```

```python
# Read / import the data
df = pd.read_csv('pie_chart_data.csv')
df
```

```python
# Simple matplotlib bar chart
xy_size = 13
plt.figure(figsize=(10, 8))

plt.pie(df['Number of Cars'],
        colors=sns.color_palette('colorblind'), 
        autopct='%.2f%%',
        textprops= {'size': 'x-large', 'fontweight': 'bold',
                    'color': 'w', 'rotation': 40})

plt.title('Cars Sold by Fuel Engine Type', fontsize=16, fontweight='bold')
plt.savefig('cars_sold_pie_graph')
plt.legend(df['Engine Fuel Type'])
plt.show()

# Note: general palette -> sns.set_palette('colorblind')
```
