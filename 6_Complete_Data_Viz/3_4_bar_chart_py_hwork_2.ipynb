##### 0. Importing libs and reading data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(pd.__version__)
print(sns.__version__)
```

```python
# Read / import the data
df_used_cars = pd.read_csv('bar_chart_homework_data.csv')
df_used_cars
```

##### 1. Using Seaborne barplot

```python
fig= plt.figure(figsize=(9, 6))
splot = sns.barplot(x=df_used_cars['Cities'],
                    y=df_used_cars['Frequency'],
                    color='midnightblue')
plt.title('Ice cream sales by Citie', fontsize=16, fontweight='bold', loc='left')
plt.bar_label(splot.containers[0], fontweight='bold')
plt.xlabel(None)
plt.ylabel('Number of sales')
plt.show()
fig.savefig('sns_bar_chart_hwork_2', bbox_inches='tight')
```
