##### 0. Importing libs and reading data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(pd.__version__)
print(sns.__version__)
```

```python
# Read / import the data
df_used_cars = pd.read_csv('bar_chart_homework_data.csv')
df_used_cars
```

##### 1. Using Seaborne barplot

```python
fig= plt.figure(figsize=(9, 4))
splot = sns.barplot(x=df_used_cars['Cities'],
                    y=df_used_cars['Frequency'],
                    color='grey')
plt.title('Cars Listings by Brand', fontsize=16, fontweight='bold', loc='left')
plt.bar_label(splot.containers[0], fontweight='bold')
plt.ylabel(None)
plt.xlabel('Number of Listings')
plt.yticks(rotation=45)
plt.show()
fig.savefig('sns_bar_chart_h', bbox_inches='tight')
```

##### 2. Using matplotlib plt.barh()

```python
# Simple matplotlib bar chart
xy_size = 13

plt.figure(figsize=(10, 5))
plt.barh(y=df_used_cars['Brand'], width=df_used_cars['Cars Listings'],
        color='mediumblue')           # color='rgbwymc'  doesn't work 
plt.title('Used Cars Listing by Brand', fontsize=16, fontweight='bold')
plt.xlabel('Listing Cars', fontsize=xy_size)
plt.yticks(rotation=45, fontsize=xy_size)
plt.xticks(fontsize=xy_size)
plt.savefig('used_cars_bar_graph_h')
plt.show()
```
