```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()          # set default plots theme, alias for .set_theme()
sns.set_theme()     # set aspects of the visual theme for all matplotlib and seaborn plots
```

Create a regression plot on the data set of car insurance in Sweden.

Decide which is your x- and which your y-axis

```python
# Import/read the data
df_regression = pd.read_csv('9_regression_auto_insurance_sweden.csv')
df_regression
```

```python
print(df_regression.info())
```

```python
# First a Seaborn scatter for first view of the data
plt.figure(figsize=(12, 8))

sns.scatterplot(x=df_regression['Claims'],       # x-axis (var 1)
                y=df_regression['Payment'],            # y-axis (var 2)
                alpha=0.8,                          # Transparency
                s=50,
                color='midnightblue')                              # Size of each point

plt.title('Relationship between Claims and Payment',
          fontsize=16, fontweight='bold')
plt.xlabel('Claims', fontweight='bold')
plt.ylabel("Payment", fontweight='bold')
plt.show()

```

### sns.regplot() -regressoin plot -; and sns.lmplot() -lineal model plot-
### 1. sns.regplot()

```python
#plt.figure(figsize=(12, 8))
sns.set(rc={'figure.figsize': (10, 7)}) # sns.set funct. w/rc parameter and figure.figsize key and value

sns.regplot(x='Claims',
            y='Payment',
            data=df_regression,
            #color='green'  # color for all the chart
            scatter_kws={'color': 'black', 'alpha': 0.6},
            line_kws={'color': 'red'})

plt.xlabel('plt.xlabel')
plt.ylabel('plt.ylabel')
plt.title('plt.title', fontdict={'size': 14, 'weight': 'bold'})
plt.show()
```

### 2. sns.lmplot

```python
# Size and aspect in lmplot is ajusted with height and aspect
sns.lmplot(x='Claims',
           y='Payment',
           data=df_regression,
           height=6,
           aspect=1.5,
           scatter_kws={'color': 'black', 'alpha': 0.7},
           line_kws={'color': 'red'})

plt.xlabel('plt.xlabel')
plt.ylabel('plt.ylabel')
plt.title('plt.title', fontdict={'size': 14, 'weight': 'bold'})
plt.show()
```

1. I decided Claims as x-axis, then we can observe the variations of the payments in function of the claims variance.
2. We can see that when the claims grow the payment grow. Is a positive linear relationship, the slope is positive.
3. Title: Relationship between Payment and Claims
> What we observe is that there seems to be a linear relationship between the two variables. The higher the insurance claim, the higher the payment it has received.
