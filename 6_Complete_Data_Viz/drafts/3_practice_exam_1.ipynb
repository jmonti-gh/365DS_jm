### Create Vertical and Horizontal Bar Charts
- Bases on the data in the grocery_store_items.csv
- Chart should display the grocery items on the x-axis and the products sold on the y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(pd.__version__)
print(sns.__version__)
```

```python
# Read / import the data
df = pd.read_csv('grocery_store_items.csv', sep=';')
df
```

```python
# Vertical Bar Chart using sns.barplot()
fig = plt.figure(figsize=(10, 5))
splot = sns.barplot(x=df['Grocery Items'], y=df['Units Sold'])
plt.bar_label(splot.containers[0])
plt.xticks(rotation=45)
plt.show()
```

```python
# Horizontal Bar Chart using sns.barplot()
fig = plt.figure(figsize=(10, 5))
splot = sns.barplot(x=df['Units Sold'], y=df['Grocery Items'])
plt.bar_label(splot.containers[0])
plt.yticks(rotation=45)
plt.show()
```
